```python
data_week = pd.read_csv('P:/TYDY/Behavioral_Weekday.csv')
data_week = data_week.drop(['Age','Gender','Department','Band','App_Logins','Web_Logins','Total_logins','Form_Completed.3','First_login','First_login.1','First_login_After_which_reminder','Whether_logged_in','Logged_or_not','Days_Worked','Months_Worked','Months_Worked_Roundoff','Mobile_OS_used','Browser_used','Form_Sent','Form_Completed','Form_Completed.1','Form_Completed.2','Total_Hours','Completion_Rate_Content','Completion_Rate_Form','Start_Date','End_Date'],axis=1)
data_week=data_week.dropna(subset=['TS1'])
data_week=data_week.reset_index(drop=True)
data_week=data_week.astype(int)
data_week
```

```python
col_= ['0','0','0','0','0','0','0']
col = pd.DataFrame(col_, columns = ['Weekday'])
col=col.astype(int)
col
```

```python
x=data_week.shape[0]
y=data_week.shape[1]-1
for i in range(0,x):
    for j in range (0,y):
        if data_week.loc[i][j] !=  data_week.loc[i][j+1]:
            col.Weekday[data_week.loc[i][j]]=col.Weekday[data_week.loc[i][j]]+1
for i in range(0,x):
    for j in range (y,y+1) :
            col.Weekday[data_week.loc[i][j]]=col.Weekday[data_week.loc[i][j]]+1
```
