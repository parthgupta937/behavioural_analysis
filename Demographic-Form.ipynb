```python
df_Age = pd.DataFrame(columns=["Age", "% of Defaulters"])
i=0
for field in list(hr5['Age'].unique()):
    ratio = hr5[(hr5['Age']==field)&(hr5['Completion_Rate_Form']==0)].shape[0] / hr5[hr5['Age']==field].shape[0]
    df_Age.loc[i] = (field, ratio*100)
    i += 1
    #print("In {}, the ratio of leavers is {:.2f}%".format(field, ratio*100))    

df_Age = df_Age.groupby(by="Age").sum()
df_Age.plot(kind='bar',title='Defaulters by Age in the company (%)')
```
